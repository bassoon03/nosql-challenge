2. Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the `BusinessTypeID` and `BusinessType` fields.

```python
# Find the BusinessTypeID for "Restaurant/Cafe/Canteen" and return only the BusinessTypeID and BusinessType fields
query = {'BusinessType' : 'Restaurant/Cafe/Canteen'}
fields = {'BusinessType' : 1, 'BusinessTypeID' : 1}
results = establishments.find(query, fields)

for d in results:
    pprint(d)

# The list containing the BusinessTypeID for "Restaurant/Cafe/Canteen" is in the next cell. When the find function is used, the query needs to
# repeated each time something is done with it. Otherwise, whatever operation one attempts to perform on a query with the find function won't
# return anything. Query repetition is done throughout the whole challenge.

```

```python
# Prints the "BusinessTypeID" for "Restaurant/Cafe/Canteen"
results = establishments.find(query, fields)
business_id = []

for result in results:
    if result["BusinessTypeID"] != '' and result["BusinessTypeID"] not in business_id:
        business_id.append(result["BusinessTypeID"])

print(business_id)
```

3. Update the new restaurant with the `BusinessTypeID` you found.

```python
# Update the new restaurant with the correct BusinessTypeID

filter = {"BusinessName" : "Penang Flavours"}
new_values = {"$set" : {"BusinessTypeID" : business_id[0]}}

establishments.update_one(filter, new_values)
```

```python
# Confirm that the new restaurant was updated

query = {'BusinessName' : 'Penang Flavours'}
fields = {'BusinessName' : 1, 'BusinessTypeID' : business_id[0]}

updated_results = establishments.find_one(query, fields)
updated_results
```

5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude

establishments.update_many({}, [{'$set': {'geocode.longitude' : {'$toDouble' : '$geocode.longitude'}}}])

establishments.update_many({}, [{'$set': {'geocode.latitude' : {'$toDouble' : '$geocode.latitude'}}}])
```

```python
# Check that the data types for longitude and latitude were change to Decimal
establishments.find_one({})
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
# Set non 1-5 Rating Values to Null

non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Change the data type from String to Integer for RatingValue

establishments.update_many({}, [{'$set':{'RatingValue':{'$toInt':'$RatingValue'}}}])
```

```python
# Check that the data type for RatingValue has been changed to Integer

establishments.find_one({})
```

```python
# Check that the coordinates and rating value are now numbers
# This step may be redundant as the data type changes were checked in previous cells. All documents can be checked here.

docs = establishments.find()

numbers_list = []

for doc in docs:
    numbers = []
    numbers.append(doc['geocode']['longitude'])
    numbers.append(doc['geocode']['latitude'])
    numbers.append(doc['RatingValue'])
    numbers_list.append(numbers)

numbers_list
```
